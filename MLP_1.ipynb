```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
tiny = pd.read_csv("tinydata.csv", index_col=0)
le = LabelEncoder()
tiny['Acceptance'] = le.fit_transform( tiny['Acceptance'] )
X = tiny.drop('Acceptance', axis=1)
y = tiny['Acceptance']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2, random_state=25, stratify=y)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
tf.random.set_seed(25)
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(2),
    tf.keras.layers.Dense(1, activation='sigmoid')])
model.compile(loss='binary_crossentropy', metrics=['acc'], optimizer=tf.keras.optimizers.SGD())
```

```python
model.get_weights()
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),verbose=2,epochs=100)
```

```python
model.get_weights()
```

```python
y_pred_prob = model.predict(X_test)
y_pred = np.where(y_pred_prob > 0.5, 1, 0)
accuracy_score(y_test, y_pred)
```
