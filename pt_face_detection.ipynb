```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
from facenet_pytorch import InceptionResnetV1
import torchvision.transforms as transforms
```

# 1. Load any pre-trained model

```python
# Load pre-trained FaceNet model (InceptionResnetV1)
# This will replace the previously defined 'model' variable.
model = InceptionResnetV1(pretrained='vggface2').eval()
```

# 2. Define the necessary image transformations

```python
# Define preprocessing for FaceNet
# FaceNet typically expects 160x160 input and normalization to [-1, 1]
# This will replace the previously defined 'preprocess' variable.
preprocess = transforms.Compose([
    transforms.Resize(160), # FaceNet input size
    transforms.CenterCrop(160),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]), # Normalize to [-1, 1]
])

print("FaceNet model (InceptionResnetV1) loaded and preprocessing updated.")
```

```python
import torch.nn.functional as F

# Flatten the tensors for cosine similarity calculation
aj1_flat = aj1.flatten()
aj2_flat = aj2.flatten()
aj3_flat = aj3.flatten()
aj4_flat = aj4.flatten()

ss1_flat = ss1.flatten()
ss2_flat = ss2.flatten()
ss3_flat = ss3.flatten()
ss4_flat = ss4.flatten()

vj1_flat = vj1.flatten()
vj2_flat = vj2.flatten()
vj3_flat = vj3.flatten()
vj4_flat = vj4.flatten()

```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
aj1_flat = aj1_flat.detach().numpy()
aj2_flat = aj2_flat.detach().numpy()
aj3_flat = aj3_flat.detach().numpy()
aj4_flat = aj4_flat.detach().numpy()
ss1_flat = ss1_flat.detach().numpy()
ss2_flat = ss2_flat.detach().numpy()
ss3_flat = ss3_flat.detach().numpy()
ss4_flat = ss4_flat.detach().numpy()
vj1_flat = vj1_flat.detach().numpy()
vj2_flat = vj2_flat.detach().numpy()
vj3_flat = vj3_flat.detach().numpy()
vj4_flat = vj4_flat.detach().numpy()
```

```python
vect_embeddings = pd.DataFrame({'aj1':aj1_flat, 'aj2':aj2_flat, 'aj3':aj3_flat, 'aj4':aj4_flat,
                                'ss1':ss1_flat, 'ss2':ss2_flat, 'ss3':ss3_flat, 'ss4':ss4_flat,
                                'vj1':vj1_flat, 'vj2':vj2_flat, 'vj3':vj3_flat, 'vj4':vj4_flat})
sims = pd.DataFrame( cosine_similarity(vect_embeddings.T), columns=['aj1','aj2','aj3','aj4','ss1','ss2','ss3','ss4',
                                                           'vj1','vj2','vj3','vj4'] )
```

```python
import gradio as gr
import torch.nn.functional as F

# Combine all flat embeddings and their names for easy iteration
all_embeddings = {
    "aj1": {'name':'Ajay Joshi', 'emb': aj1_flat},
    "aj2": {'name':'Ajay Joshi', 'emb': aj2_flat},
    "aj3": {'name':'Ajay Joshi', 'emb': aj3_flat},
    "aj4": {'name':'Ajay Joshi', 'emb': aj4_flat},
    "ss1": {'name':'Sanjay Sane', 'emb': ss1_flat},
    "ss2": {'name':'Sanjay Sane', 'emb': ss2_flat},
    "ss3": {'name':'Sanjay Sane', 'emb': ss3_flat},
    "ss4": {'name':'Sanjay Sane', 'emb': ss4_flat},
    "vj1": {'name':'Vidyadhar Joshi', 'emb': vj1_flat},
    "vj2": {'name':'Vidyadhar Joshi', 'emb': vj2_flat},
    "vj3": {'name':'Vidyadhar Joshi', 'emb':vj3_flat},
    "vj4": {'name':'Vidyadhar Joshi', 'emb': vj4_flat}
}

```

```python
import torch.nn.functional as F

def process_webcam_image(image):
    if image is None:
        return "Please provide an image."

    # Preprocess the input image
    input_tensor = preprocess(Image.fromarray(image)).unsqueeze(0)

    # Generate embedding for the new image
    with torch.no_grad():
        new_embedding = model(input_tensor).flatten().numpy()

    results = {}
    for key, stored_data in all_embeddings.items():
        person_name = stored_data['name'] # Get the name
        stored_embedding = stored_data['emb'] # Get the embedding

        # Calculate cosine similarity
        similarity = F.cosine_similarity(torch.from_numpy(new_embedding).unsqueeze(0), torch.from_numpy(stored_embedding).unsqueeze(0))
        results[f"{person_name} ({key})"] = f"{similarity.item():.4f}"

    return results

# Create Gradio interface
iface = gr.Interface(
    fn=process_webcam_image,
    inputs=gr.Image(type="numpy", label="Capture Image from Webcam"),
    outputs=gr.Label(),
    title="Face Similarity Checker",
    description="Capture an image from your webcam and see its cosine similarity with pre-trained face embeddings."
)

# Launch the interface
iface.launch()
```
