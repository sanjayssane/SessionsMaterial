```python
from transformers import AutoModelForSeq2SeqLM
from transformers import AutoTokenizer
from transformers import GenerationConfig
import pandas as pd
```

```python
model_name='google/flan-t5-base'
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
sentence = "I am planning to buy a house valued at approximately 90 lakhs. I currently have a \
    savings of 25 lakhs. My bank is willing to provide a loan at an interest rate of 8% per annum \
    for a maximum period of 25 years. The bank stipulates a 10% down payment. I can manage a monthly EMI \
    of 70,000 INR. I also tend to invest mainly in govt. bonds and FD."

sentence_encoded = tokenizer(sentence, return_tensors='pt')
```

```python
summary = {'Intention to Buy':'House', 'Cost':'Rs. 90 lakhs'}
```

```python
sentences2 = "Hi, I'm looking to purchase a bike priced around 5 lakhs. I have saved up 1 lakhs, and my bank is\
     offering a loan at an 10.5% annual interest rate, with a maximum tenure of 5 years. The bank requires a 10% down payment.\
          I can afford to pay 20,000 INR per month as EMI. Additionally, I primarily invest in small cap index funds."
```

```python
prompt = f"""
{sentence}

What is the case?
{summary}

{sentences2}
What is the case?
"""
```

```python
print(prompt)
```

```python
inputs = tokenizer(prompt, return_tensors='pt')
output = tokenizer.decode(
    model.generate(
        inputs["input_ids"],
        max_new_tokens=50,
    )[0], 
    skip_special_tokens=True
)
print(output)
```
